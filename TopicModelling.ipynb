```python
# importing basic libraries
import pandas as pd
import numpy as np
import nltk
```

```python
# loading data:

# setting path for the file
path='/home/suraj/topic.txt'

# reading data from the txt file and storing result in documnets: 
with open(path) as f:
    documents = f.readlines()
documents = [x.strip().strip('\n').strip("'") for x in documents] 
#documents

# changing it into dataframe
df=pd.DataFrame(documents)
df.replace('', np.nan, inplace=True)
df.dropna(inplace=True)
df=pd.DataFrame(df)
df = df.reset_index(drop=True)
df
```

```python
# word2vec for word simmilarity

#import genism
    
```

```python
# cleaning th data 

tokens=[]
import re
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
stop_words=set(stopwords.words('english'))
stop=['o']
for i in range(0,len(df)):
    word_tokens = re.sub('[^a-zA-Z]', ' ', df[0][i])
    word_tokens = word_tokens.lower()
    word_tokens= word_tokens.split()
    word_tokens=[word for word in word_tokens if word not in stop_words]
    word_tokens= ' '.join(word_tokens)
    tokens.append(word_tokens)
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
results2 = [line.split() for line in tokens]
lemma_list_of_words = []

# Snowballstemmer
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

for i in range(0, len(results2)):
    l1 = results2[i]
    
    l3=' '.join([stemmer.stem(lemmatizer.lemmatize(word)) for word in l1])
    lemma_list_of_words.append(l3)
```

```python
def display_topics(H, W, feature_names, documents, no_top_words, no_top_documents):
    for topic_idx, topic in enumerate(H):
        print ("Topic %d:" % (topic_idx))
        print (" ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
        top_doc_indices = np.argsort( W[:,topic_idx] )[::-1][0:no_top_documents]
        for doc_index in top_doc_indices:
            #documents.apply(lambda x: str(TextBlob(x).correct()))
            print (documents[doc_index])
```

```python
# Performing topic modelling using LDA

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(analyzer='word',min_df=2,stop_words='english',lowercase=True,token_pattern='[a-zA-z0-9]{3,}')
#print(cv)
data_Vectorized=cv.fit_transform(lemma_list_of_words)
data_Vectorized
df = pd.DataFrame(data_Vectorized.toarray(), columns=cv.get_feature_names())
print(df.shape)

from sklearn.decomposition import LatentDirichletAllocation, TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer

lda_model = LatentDirichletAllocation(n_topics=5,              # Number of topics
                                      max_iter=5000,              # Max learning iterations
                                      learning_method='online',   
                                      random_state=100,          # Random state
                                      batch_size=15,            # n docs in each learning iter
                                      evaluate_every = -1,       # compute perplexity every n iters, default: Don't
                                      n_jobs = -1,)  

lda_output = lda_model.fit(data_Vectorized)
#print(lda_output.components_)

feature_names = cv.get_feature_names()
feature_names
n=len(feature_names)

for topic_idx, topic in enumerate(lda_output.components_):    
    print ("Topic %d:" % (topic_idx))
    print (" ".join([feature_names[i] for i in topic.argsort()[:-50 - 1:-1]]))
    
```

```python
# performimg NMF ( Non-Negeative Matrix Factorization)

no_features = n

# NMF is able to use tf-idf
tfidf_vectorizer = TfidfVectorizer(max_df=0.90, min_df=2, max_features=no_features, stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(lemma_list_of_words)
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
from sklearn.decomposition import NMF

no_topics = 5
nmf = NMF(n_components=no_topics, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf)
nmf_W = nmf.transform(tfidf)
nmf_H = nmf.components_

no_top_words = 15
no_top_documents=5
display_topics(nmf_H, nmf_W, tfidf_feature_names, lemma_list_of_words, no_top_words, no_top_documents)
```
